```python
import numpy as np
import matplotlib.pyplot as plt

# Función que implementa el método de Euler
def euler_method(f, y0, t0, t_end, h):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
    return t_values, y_values

# Función que implementa el método de Taylor de segundo orden
def taylor_method(f, dfdt, dfdy, y0, t0, t_end, h):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    for i in range(1, len(t_values)):
        t = t_values[i-1]
        y = y_values[i-1]
        y_values[i] = y + h * f(t, y) + (h**2 / 2) * (dfdt(t, y) + dfdy(t, y) * f(t, y))
    return t_values, y_values

# Ejercicio 1
# a) y' = te^3t - 2y, y(0) = 0, h = 0.5
def f1a(t, y):
    return t * np.exp(3 * t) - 2 * y

t1a, y1a = euler_method(f1a, y0=0, t0=0, t_end=1, h=0.5)

print("Solución Ejercicio 1a:")
for t, y in zip(t1a, y1a):
    print(f"t = {t:.2f}, y = {y:.6f}")


# b) y' = 1 + (t - y)^2, y(2) = 1, h = 0.5
def f1b(t, y):
    return 1 + (t - y)**2

t1b, y1b = euler_method(f1b, y0=1, t0=2, t_end=3, h=0.5)

print("Solución Ejercicio 1b:")
for t, y in zip(t1b, y1b):
    print(f"t = {t:.2f}, y = {y:.6f}")

# c) y' = 1 + y/t, y(1) = 2, h = 0.25
def f1c(t, y):
    return 1 + y/t

t1c, y1c = euler_method(f1c, y0=2, t0=1, t_end=3, h=0.25)


print("Solución Ejercicio 1c:")
for t, y in zip(t1c, y1c):
    print(f"t = {t:.2f}, y = {y:.6f}")

# d) y' = cos(2t) + sen(3t), y(0) = 1, h = 0.25
def f1d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

t1d, y1d = euler_method(f1d, y0=1, t0=0, t_end=1, h=0.25)

print("Solución Ejercicio 1b:")
for t, y in zip(t1d, y1d):
    print(f"t = {t:.2f}, y = {y:.6f}")

# Ejercicio 2
# Las soluciones reales están dadas en la imagen, puedes comparar los valores obtenidos con los valores analíticos.

# Ejercicio 3
# a) y' = y/t - (y/t)^2, y(1) = 1, h = 0.1
def f3a(t, y):
    return y/t - (y/t)**2

t3a, y3a = euler_method(f3a, y0=1, t0=1, t_end=2, h=0.1)

print("Solución Ejercicio 3a:")
for t, y in zip(t3a, y3a):
    print(f"t = {t:.2f}, y = {y:.6f}")

# b) y' = 1 + y/t + (y/t)^2, y(1) = 0, h = 0.2
def f3b(t, y):
    return 1 + y/t + (y/t)**2

t3b, y3b = euler_method(f3b, y0=0, t0=1, t_end=3, h=0.2)

print("Solución Ejercicio 3b:")
for t, y in zip(t3b, y3b):
    print(f"t = {t:.2f}, y = {y:.6f}")

# c) y' = -(y + 1)(y + 3), y(0) = -2, h = 0.2
def f3c(t, y):
    return -(y + 1) * (y + 3)

t3c, y3c = euler_method(f3c, y0=-2, t0=0, t_end=2, h=0.2)

print("Solución Ejercicio 3c:")
for t, y in zip(t3c, y3c):
    print(f"t = {t:.2f}, y = {y:.6f}")

# d) y' = -5y + 5t^2 + 2t, y(0) = 1/3, h = 0.1
def f3d(t, y):
    return -5 * y + 5 * t**2 + 2 * t

t3d, y3d = euler_method(f3d, y0=1/3, t0=0, t_end=1, h=0.1)

print("Solución Ejercicio 3d:")
for t, y in zip(t3d, y3d):
    print(f"t = {t:.2f}, y = {y:.6f}")

# Ejercicio 4
# Las soluciones reales están dadas en la imagen, puedes comparar los valores obtenidos con los valores analíticos.

# Ejercicio 5
# Interpolación lineal y comparación de valores específicos
def interpolate_linear(t_values, y_values, t):
    return np.interp(t, t_values, y_values)

interp_y1 = interpolate_linear(t3a, y3a, 0.25)
interp_y2 = interpolate_linear(t3d, y3d, 0.54)
interp_y3 = interpolate_linear(t3a, y3a, 1.25)
interp_y4 = interpolate_linear(t3d, y3d, 1.93)

t_values_3a, y_values_3a = euler_method(f3a, y0=1, t0=1, t_end=2, h=0.1)
t_values_3d, y_values_3d = euler_method(f3d, y0=1/3, t0=0, t_end=1, h=0.1)

# Puntos para interpolación lineal
t_interp_3a = [0.25, 1.25]
t_interp_3d = [0.54, 1.93]

# Valores interpolados
y_interp_3a = [np.interp(t, t_values_3a, y_values_3a) for t in t_interp_3a]
y_interp_3d = [np.interp(t, t_values_3d, y_values_3d) for t in t_interp_3d]

# Gráfica de los resultados de interpolación para 3a y 3d
plt.figure(figsize=(12, 6))

# Gráfico para la solución aproximada de y(t) del inciso a) del ejercicio 3
plt.subplot(1, 2, 1)
plt.plot(t_values_3a, y_values_3a, 'o-', label="Aproximación Euler")
plt.plot(t_interp_3a, y_interp_3a, 'x', color='red', label="Valores interpolados")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Ejercicio 5 - Interpolación en Ejercicio 3a")
plt.legend()

# Gráfico para la solución aproximada de y(t) del inciso d) del ejercicio 3
plt.subplot(1, 2, 2)
plt.plot(t_values_3d, y_values_3d, 'o-', label="Aproximación Euler")
plt.plot(t_interp_3d, y_interp_3d, 'x', color='red', label="Valores interpolados")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Ejercicio 5 - Interpolación en Ejercicio 3d")
plt.legend()

plt.tight_layout()
plt.show()

# Ejercicio 6
# Usar el método de Taylor de orden 2
# a) y' = te^3t - 2y, y(0) = 0, h = 0.5
def dfdt1a(t, y):
    return np.exp(3 * t) + 3 * t * np.exp(3 * t)

def dfdy1a(t, y):
    return -2

t6a, y6a = taylor_method(f1a, dfdt1a, dfdy1a, y0=0, t0=0, t_end=1, h=0.5)

# Otros incisos del ejercicio 6 se resuelven de forma similar cambiando las funciones.

# Gráfica de resultados (opcional)
plt.plot(t1a, y1a, label="Euler: Ejercicio 1a")
plt.plot(t6a, y6a, label="Taylor: Ejercicio 6a")
plt.grid()
# Agregar las demás soluciones si es necesario
plt.legend()
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Resultados de los métodos numéricos")
plt.show()
```
